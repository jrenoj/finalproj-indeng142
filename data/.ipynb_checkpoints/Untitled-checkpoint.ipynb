```python
import pandas as pd
df = pd.read_csv("C:/Users/rebek/finalproj-indeng142/data/student-mat.csv", sep=";")
print(df.head())  # Preview the first few rows
```

```python
print(df.select_dtypes(include=['object']).head())  # Show sample non-numeric data
print(df.select_dtypes(include=['object']).columns)  # List categorical columns
```

```python
df = pd.get_dummies(df, drop_first=True)  # One-hot encode categorical variables
print(df.head())  # Preview transformed data
print(df.dtypes)  # Verify all columns are numeric
```

```python
df = df.astype(int)  # Convert boolean values to integer!
```

```python
X = df.drop("G3", axis=1)  # all the features
y = df["G3"]  # target variable
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)  # Add intercept
print("New Shape of X:", X.shape) 
print("Updated Feature List:", X.columns.tolist())  # Confirm "const" is included
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_vif = X.copy() 
vif_data = pd.DataFrame()
vif_data["Feature"] = list(X_vif.columns)  # the index matches column names
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print(vif_data.sort_values(by="VIF", ascending=False))  
```

```python
X_filtered = X.drop(["Fjob_other", "Fjob_services"], axis=1).copy()  # Dropping high VIF features
print("New Shape of X_filtered:", X_filtered.shape) 
```

```python
X_filtered = X_filtered.reset_index(drop=True)  # Aligning index
```

```python
vif_data_filtered = pd.DataFrame()
vif_data_filtered["Feature"] = X_filtered.columns  # Ensure index matches column names
vif_data_filtered["VIF"] = [variance_inflation_factor(X_filtered.values, i) for i in range(X_filtered.shape[1])]

print(vif_data_filtered.sort_values(by="VIF", ascending=False)) #successfully reduced MULTICOLINEARITY by removing features with high VIF
```

```python
#checking for highly correlated features now using other method
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = X_filtered.corr().abs()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", linewidths=0.5)
plt.show()

# highly correlated feature pairs (>0.8)USING THIS as the threshold point 
high_corr_features = [column for column in corr_matrix.columns if any(corr_matrix[column] > 0.8)]
print("Highly correlated features:", high_corr_features) #gather all the highly correlated features
```

```python
features_to_drop = ["G1", "Dalc", "Medu", "Mjob_other", "guardian_mother"]  # keeping one feature in each correlated pair, one should be sufficient enough according to lecture
X_refined = X_filtered.drop(features_to_drop, axis=1)
print("New Shape of X_refined:", X_refined.shape)  # Confirm feature count reduction
```

```python
#Running again
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data_refined = pd.DataFrame()
vif_data_refined["Feature"] = X_refined.columns
vif_data_refined["VIF"] = [variance_inflation_factor(X_refined.values, i) for i in range(X_refined.shape[1])]

print(vif_data_refined.sort_values(by="VIF", ascending=False))  # Verify multicollinearity improvement
```

```python
# Assess p-values and model output
model_refined = sm.OLS(y, X_refined).fit()
print(model_refined.summary())
```

```python
insignificant_features = model_refined.pvalues[model_refined.pvalues > 0.05].index.tolist() #gathering features at the insignificant level
print( insignificant_features)#features to consider dropping, but almsot ALL the features seem to have extreme pvalues for some reason 
```

```python
ols_predictions = model_refined.predict(X_refined)
from sklearn.metrics import mean_squared_error
mse_ols = mean_squared_error(y, ols_predictions)
print("Mean Squared Error (MSE) for OLS:", mse_ols)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_refined) 
print("Shape of X_scaled:", X_scaled.shape)
```

```python
#ridge
ridge = Ridge(alpha=0.1)  # Regularization strength (using class example for this parameter)
ridge.fit(X_scaled, y)
print("Ridge coefficients:", ridge.coef_)  #to see adjustments
```

```python
predictions_ridge = ridge.predict(X_scaled)
mse_ridge = mean_squared_error(y, predictions_ridge)
print("Mean Squared Error (MSE) with Ridge:", mse_ridge)  # Compare to OLS MSE
```
